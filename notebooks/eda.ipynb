```python
import pandas as pd

#TAREA
#pivotear columnas region, planta, precipitacion. El valor es precipitacion

filename ='chirps.csv'
colnames =["ch_fecha","ch_region","ch_planta","ch_embalse","ch_precipitacion"]
dfchips = pd.read_csv(filename, names=colnames, header=0)
dfchips.head()
```

```python
df_pt_embalse = pd.pivot(dfchips, index='ch_fecha', columns='ch_embalse', values='ch_precipitacion')
df_pt_embalse.head()
```

```python
sr_group_planta = dfchips.groupby(["ch_fecha", "ch_planta"])["ch_precipitacion"].mean()
df_group_planta = sr_group_planta.reset_index()
df_pt_planta = pd.pivot(df_group_planta, index='ch_fecha', columns='ch_planta', values='ch_precipitacion')
df_pt_planta.head()
```

```python
sr_group_region = dfchips.groupby(["ch_fecha", "ch_region"])["ch_precipitacion"].mean()
df_group_region = sr_group_region.reset_index()
df_pt_region = pd.pivot(df_group_region, index='ch_fecha', columns='ch_region', values='ch_precipitacion')
df_pt_region.head()
```
